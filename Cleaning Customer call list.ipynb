```python
import pandas as pd
df = pd.read_excel(r'C:\Users\user\Documents\web scrape data\Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
#remove the duplicates
df.drop('Not_Useful_Column', axis=1, inplace=True)
#remove column not needed
df['Last_Name'] = df['Last_Name'].str.strip('._/')
df
```

```python
df['Phone_Number'] = (
    df['Phone_Number'].str.replace('[^A-Za-z0-9]', '', regex=True)  # Remove non-alphanumeric characters
    .apply(lambda x: str(x))  # Convert to string
    .apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] if len(x) >= 10 else x)  # Format phone numbers
    .str.replace('nan', '')  # Replace 'nan--' with ''
    .str.replace('Na', '')  # Replace 'Na--' with ''
)
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',n= 2,expand=True)
#n=2 signifies the number of commas present to split on
#n=2 since there is a maximum of 2 commas
df['Paying Customer']= df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer']= df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('No', 'N')
df = df.replace('N/a','')
#no.str added because we are not looking at columns specfically
df
```

```python
df = df.fillna('')
df = df[df['Do_Not_Contact'] != 'Y']
# only return/need customers that can contacted
#so where do_not_contact = Y should be removed
```

```python
df = df[df['Phone_Number'] != '']
#return only customers with phone numbers
```

```python
import csv

header = df.columns
data = df.values

type(data)

#nweline = '', when we insert the data it dosent have a space between each CSV
with open('Customer call.csv', 'w', newline='', encoding = 'UTF8') as file:
    writer = csv.writer(file)#creating the csv
    writer.writerow(header)#insert the header
    writer.writerows(data)#insert the data
```
